```python
import numpy as np 
import matplotlib.pyplot as plt 
import librosa
import librosa.display 
import pickle
import time
```

```python
def plotSpectrogram(audio: np.ndarray, sample_rate: int = 4800):
    '''
    * Plot melspectrogram 
    '''
    audio = np.array(audio.reshape(-1, 128, 188), dtype = "float64")
    
    random = np.random.randint(len(audio))
    random_audio = audio[random]
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(librosa.power_to_db(random_audio,
                                            ref=np.max),
                         y_axis='mel', fmax = 2000,
                          x_axis='time')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Mel spectrogram')
    plt.tight_layout()
```

```python
audio = pickle.load(open("extracted_audio/gibbon_spectrogram/D1-01-01b-20200929_032722.Table.1.selections.txt.pkl", "rb"))
```

```python
plotSpectrogram(audio)
```
